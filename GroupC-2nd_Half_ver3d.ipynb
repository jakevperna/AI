```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from sklearn.preprocessing import MinMaxScaler

# Load CSV file into a Pandas dataframe
df= pd.read_excel('Figure 12 Gulf countries exports by product sample data set, WTO.xlsx',skiprows = [0,1])
df.fillna(1)
df = df.iloc[90:108, [1, 55,56,57,58,59,60,61,62,63,64,65,66,67,68,69,70,71,72,73,74,75]]
df = df.melt(id_vars='Product/Sector', var_name="Year", value_name='trade')

# Convert the dataframe to a numpy array
melt = df

# Split the data into training and testing sets
melt['Year'] = melt['Year'].str.extract('(\d+)', expand=False).astype(int)
split_point = 2015
train_data = melt[melt['Year'] < split_point].copy()
test_data = melt[melt['Year'] >= split_point].copy()
#train_data = data[:-10]
#test_data = data[-10:]


# Normalize the data between 0 and 1
scaler = MinMaxScaler()
train_data = scaler.fit_transform(train_data)
test_data = scaler.transform(test_data)
#train_data = train_data.reshape(-1, 1)
#test_data = test_data.reshape(-1, 1)


# Define the number of time steps and features
time_steps = 10
n_features = 1

# Create sequences of data for the model
def create_sequences(data, time_steps):
    X, y = [], []
    for i in range(len(data)-time_steps):
        X.append(data[i:i+time_steps])
        y.append(data[i+time_steps])
    return np.array(X), np.array(y)

X_train, y_train = create_sequences(train_data, time_steps)
X_test, y_test = create_sequences(test_data, time_steps)

# Define the deep learning model architecture
model = Sequential()
model.add(LSTM(64, input_shape=(time_steps, n_features), return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(32, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(16))
model.add(Dense(1))

# Compile the model
model.compile(loss='mse', optimizer='adam')

# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))

# Predict future exports
future_data = np.array([[100], [120], [130], [140], [150], [160], [170], [180], [190], [200]])
future_data_scaled = scaler.transform(future_data)
X_future, y_future = create_sequences(future_data_scaled, time_steps)
predictions = model.predict(X_future)
predictions_scaled = scaler.inverse_transform(predictions)

# Print the predicted future exports
print(predictions_scaled)
```
